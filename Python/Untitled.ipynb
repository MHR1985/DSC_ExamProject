```python
import cv2 # workswith images 
from matplotlib import pyplot # for visualisation
import numpy as np
```

```python
img = cv2.imread('../Data/dataset_B_Facialimages/closedface/closed_eye_0033.jpg_face_2.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#gray = img.astype(np.int16)
eye_cascade=cv2.CascadeClassifier('../Data/haarcascade_eyes.xml')
face_cascade = cv2.CascadeClassifier('../Data/haarcascade_face.xml')
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
myimage(img)
```

```python
for (x,y,w,h) in faces:
     cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
     roi_gray = gray[y:y+h, x:x+w]
     roi_color = img[y:y+h, x:x+w]
     eyes = eye_cascade.detectMultiScale(roi_gray)
     for (ex,ey,ew,eh) in eyes:
         cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
myimage(img)
```

```python
def myimage(image):
    # image view
    pyplot.imshow(image, cmap='gray', vmin=0, vmax=255)
    pyplot.show()
    # pixel view
    print('image size: ', image.shape)
    ('pixel matrix:\n', image)
```
